### Imports

```python
import pandas as pd
import numpy as np
```

### Define variables

```python
train_path = "../../Data/raw/Modelar_UH2020.txt"
test_path = "../../Data/raw/Estimar_UH2020.txt"
store_path = "../../Data/raw_pkl/"
```

### Read txt

```python
def readRAW():
    X_train = pd.read_csv(train_path, sep="|", header=0)
    Y_train = X_train['CLASE']
    X_train = X_train.drop(columns='CLASE')
    X_test = pd.read_csv(test_path, sep="|", header=0)
    return X_train, Y_train, X_test
```

### Analyze data

```python
X_train , Y_train, Y_test = readRAW()
```
