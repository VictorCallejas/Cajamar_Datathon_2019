### Imports

```python
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.6f' % x)

import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import display_html 

import sys; sys.path.insert(0, '..')
import lib
```

### Read Data

```python
X_train, y_train = lib.readHDF('../../Data/raw/hdf5/data.hdf5')
X_test = lib.readHDF('../../Data/raw/hdf5/data.hdf5',estimar=True)
```

### Analyze distributions

#### Modelar Distribution

```python
X_train_obj = X_train['CADASTRALQUALITYID']
X_train = X_train.drop(columns={'ID','X','Y','CADASTRALQUALITYID'});
```

```python
X_test_obj = X_test['CADASTRALQUALITYID']
X_test = X_test.drop(columns={'ID','X','Y','CADASTRALQUALITYID'});
```

```python
X_train_desc = X_train.describe()
X_train_desc = X_train_desc.drop('count')
X_train_desc
```

#### Estimar Distribution

```python
X_test_desc = X_test.describe()
X_test_desc = X_test_desc.drop('count')
X_test_desc
```

#### Diferencia entre las distribuciones de Modelar y Estimar

```python
X_diff_desc = X_train_desc - X_test_desc
X_diff_desc = X_diff_desc.abs()
X_diff_desc
```

Se observa como las distribuciones de las variables son las mismas menos en el caso de la variable area

```python
tmp = pd.concat([X_train_desc['AREA'],X_test_desc['AREA'],X_diff_desc['AREA']],axis=1)
tmp.columns = ['Train','Test','Diff']
tmp = tmp.round().astype(int)
tmp
```

La media de la variable área de 525m^2 superior en el test set, más del doble.

### Area en contexto

```python
XY_train = pd.DataFrame({
    'AREA':X_train['AREA'],
    'CLASE':y_train
})
XY_train.CLASE = XY_train.CLASE.apply(str)
area_clase = XY_train.groupby('CLASE')['AREA'].mean().round().astype(int)
area_clase
```

Medias de área según clase
