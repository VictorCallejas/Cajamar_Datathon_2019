### Imports

```python
import pandas as pd
pd.set_option('display.expand_frame_repr', False)

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

import sys; sys.path.insert(0, '..')
import lib
```

### Read data

```python
X, y = lib.readHDF('../../Data/raw/hdf5/data.hdf5')
```

```python
X['CLASE']=y
y.value_counts()
X.drop(columns=['X','Y','ID'],inplace=True)

#X.dropna(inplace=True)

X = lib.cadastralQualityToNumeric(X)
X = lib.buildingFloorToNumeric(X)
#X = lib.sumSatChannels(X)
```

```python
colnames = list(X.columns)
colnames.remove('CLASE')
```

### Plot

Se hace un violinplot para cada variable y con las clases separadas

```python
g = X.groupby('CLASE')

for var in colnames:
    #tmp = X[var].between(X[var].quantile(.05), X[var].quantile(.95)) # without outliers
    print('VARIABLE: ', var)
    print(g[var].describe().drop(columns=['count']))
    ax = plt.figure(figsize=(15,6))
    ax = sns.violinplot(x="CLASE", y=X[var], data=X)
    plt.show()
```
